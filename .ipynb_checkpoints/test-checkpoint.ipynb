```python
import numpy as np
from scipy import stats
import plotly.graph_objects as go

def plot_3d_kde(data_3d, grid_points=20, threshold = 0):
    # Convert 3D array indices to coordinates
    grid_x, grid_y, grid_z = np.meshgrid(
        np.arange(data_3d.shape[0]),
        np.arange(data_3d.shape[1]),
        np.arange(data_3d.shape[2]),
        indexing='ij'
    )
    
    # Create interactive 3D plot
    fig = go.Figure(data=go.Volume(
        x=grid_x.flatten(),
        y=grid_y.flatten(),
        z=grid_z.flatten(),
        value=data_3d.flatten(),
        isomin=threshold,
        isomax=data_3d.max(),
        opacity=0.1,  # Needs to be small to see through all surfaces
        surface_count=20,  # Number of isosurfaces, more surfaces create smoother rendering
        colorscale='RdYlBu'  # You can change this to any plotly colorscale
))

    # Update layout
    fig.update_layout(
        scene=dict(
            xaxis_title='X',
            yaxis_title='Y',
            zaxis_title='Z',
            camera=dict(
                up=dict(x=0, y=0, z=1),
                center=dict(x=0, y=0, z=0),
                eye=dict(x=1.5, y=1.5, z=1.5)
            )
        ),
        width=800,
        height=800,
        title='3D KDE Plot'
    )
    
    # Add opacity slider
    fig.update_layout(
        sliders=[dict(
            active=3,
            currentvalue={"prefix": "Opacity: "},
            pad={"t": 50},
            steps=[dict(
                method="update",
                args=[{"opacity": [i/10]}],
                label=str(i/10)
            ) for i in range(1, 11)]
        )]
    )

    return fig
```

```python
plot_3d_kde(data.squeeze(), threshold=0.01)
```

```python
import torch
import torch.nn.functional as F
import numpy as np

def augment_3d_data(image_tensor, num_augmentations=100):
    """
    Augment a single 3D image tensor by applying translations and 90-degree rotations.
    """    
    # Create a list to store augmented images
    augmented_images = [image_tensor]
    
    # Get dimensions
    _, c, d, h, w = image_tensor.shape
    
    for i in range(num_augmentations):
        # Create a copy of the original
        aug_img = image_tensor.clone()
        
        # Randomly decide augmentation type
        aug_type = np.random.choice(['translation', 'rotation', 'both'])
        
        # Apply translation
        if aug_type in ['translation', 'both']:
            # Random translation (-5 to 5 pixels in each dimension)
            shift_d = np.random.randint(-5, 6)
            shift_h = np.random.randint(-5, 6)
            shift_w = np.random.randint(-5, 6)
            
            # Use grid_sample for smooth translation
            grid_d, grid_h, grid_w = torch.meshgrid(
                torch.linspace(-1, 1, d),
                torch.linspace(-1, 1, h),
                torch.linspace(-1, 1, w),
                indexing='ij'
            )
            
            # Shift the grid
            grid_d = grid_d - (2 * shift_d / d)
            grid_h = grid_h - (2 * shift_h / h)
            grid_w = grid_w - (2 * shift_w / w)
            
            # Stack the grid
            grid = torch.stack([grid_d, grid_h, grid_w], dim=-1).unsqueeze(0)
            
            # Apply transformation
            aug_img = F.grid_sample(aug_img, grid, align_corners=True, mode='bilinear')

        
        # Apply rotation (90-degree increments around random axis)
        if aug_type in ['rotation', 'both']:
            # Choose a random axis and number of 90-degree rotations
            axis = np.random.randint(0, 3)  # 0=depth, 1=height, 2=width
            k = np.random.randint(1, 4)  # 1, 2, or 3 times 90 degrees
            
            # Rotate using torch.rot90
            if axis == 0:  # Rotate around depth axis
                aug_img = torch.rot90(aug_img, k, dims=[3, 4])
            elif axis == 1:  # Rotate around height axis
                aug_img = torch.rot90(aug_img, k, dims=[2, 4])
            else:  # Rotate around width axis
                aug_img = torch.rot90(aug_img, k, dims=[2, 3])
        
        augmented_images.append(aug_img)
    
    # Concatenate all augmented images
    augmented_dataset = torch.cat(augmented_images, dim=0)
    
    return augmented_dataset

def display_augmentation_info(augmented_dataset):
    """
    Display information about the augmented dataset.
    """
    print(f"Augmented dataset shape: {augmented_dataset.shape}")
    print(f"Number of unique samples: {augmented_dataset.shape[0]}")
    print(f"Min value: {augmented_dataset.min().item()}")
    print(f"Max value: {augmented_dataset.max().item()}")
    print(f"Mean value: {augmented_dataset.mean().item()}")

# Example usage
if __name__ == "__main__":
    # Create a sample 3D tensor with shape [1, 1, 44, 44, 44]
    # In a real scenario, this would be your actual data
    original_tensor = torch.tensor(np.load('datasets/Fe_super.npy'), dtype=torch.float32).unsqueeze(0)
    
    # Apply augmentation to create 20 new samples
    augmented_dataset = augment_3d_data(original_tensor, num_augmentations=100)
    
    # Display information about the augmented dataset
    display_augmentation_info(augmented_dataset)
    
    # Now you can use this augmented dataset for training
    # example_dataset = torch.utils.data.TensorDataset(augmented_dataset)
    # example_loader = torch.utils.data.DataLoader(example_dataset, batch_size=4, shuffle=True)
```
